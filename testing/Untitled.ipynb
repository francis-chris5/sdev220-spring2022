```python
import numpy, time
import matplotlib.pyplot as plt
from tqdm import *
```

```python
X = []
y = []
for i in tqdm(range(100)):
    X.append(i)
    y.append(numpy.random.rand())
    time.sleep(0.1) ## the sleep is just to see the progress bar move
print("Okay, ready to plot now...")
```

```python
plt.plot(X, y)
```
